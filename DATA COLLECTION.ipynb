```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service as EdgeService
from selenium.webdriver.edge.options import Options as EdgeOptions
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
import re
```

```python
title = []
Kms_driven = []
Petrol_Diesel=[]
Car_Type=[]
Car_owner_Type=[]
Price=[]
d = {"mumbai":2378,"hyderabad":3686,"new-delhi":2,"bangalore":4709,"kolkata":777,"chennai":5732,"pune":2423,"kochi":6356,"ahmedabad":1692,"lucknow":290,"jaipur":2130,"vadodara":1634,"agra":136,"rajkot":1606,"chandigarh":769,"surat":1605,"gurgaon":5,"noida":134,"ghaziabad":132,"nashik":2598,"nagpur":2713,"coimbatore":6105,"indore":2920,"patna":8184,"ludhiana":666,"faridabad":6,"mohali":650,"zirakpur":648,"panchkula":79,"bhubaneswar":7323,"calicut":6289,"guwahati":6772,"jodhpur":2310,"madurai":5951,"mangalore":4862,"udaipur":2185,"vijayawada":3712,"visakhapatnam":4408,"raipur":3460,"ajmer":2137,"allahabad":213,"alwar":2115,"amritsar":684,"bhopal":2987,"mysore":4747,"meerut":404,"kota":2234,"hisar":40,"ranchi":8518,"salem":6059,"bareiily":359}
for a,b in d.items():
    
    service = EdgeService(EdgeChromiumDriverManager().install())
    options = EdgeOptions()
    driver = webdriver.Edge(service=service, options=options)
    
    url = f"https://www.cars24.com/buy-used-cars-{a}/?sort=bestmatch&serveWarrantyCount=true&storeCityId={b}"
    print(url)
    #request_header = {'Content-Type': 'text/html; charset=UTF-8','User-Agent': 'Chrome/101.0.0.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0','Accept-Encoding': 'gzip, deflate, br'}
    #page = requests.get(url,headers = request_header)
    #print(page.text)
    #soup_1 = BeautifulSoup(page.text)
    driver.get(url)
    html = driver.page_source
    soup_1 = BeautifulSoup(html, "html.parser")
    for i in soup_1.find_all("div",class_="styles_contentWrap__9oSrl"):
       
        text=i.text
        t = i.find("span",class_ ="sc-braxZu kjFjan")
        if t:
            title.append(t.text)
        else:
            title.append(np.nan)
            #print(text)
        kms_driven = re.findall(r".(\d\d\W.\dk\skm)",text)   
        if kms_driven:
                Kms_driven.append(kms_driven[0])
        else:
          Kms_driven.append(np.nan)
        P_D = re.findall(r"km(Diesel|Petrol|CNG)",text)  
        if P_D:
            Petrol_Diesel.append(P_D[0])
        else:
          Petrol_Diesel.append(np.nan)
        cartype = re.findall(r"(Auto|Manual)",text)   
        if cartype:
            Car_Type.append(cartype[0])
        else:
          Car_Type.append(np.nan)
        car_owner = re.findall(r"(1st|2nd)(\sowner)",text)   
        if car_owner:
            Car_owner_Type.append(car_owner[0])
        else:
          Car_owner_Type.append(np.nan)
        r1= re.findall(r"m(₹.+lakh)",text)   
        if r1:
            Price.append(r1[0])
        else:
            Price.append(np.nan)
df= pd.DataFrame({"Title":title,"Kms_driven":Kms_driven,"Petrol_Diesel":Petrol_Diesel,"Car_Type":Car_Type,"Car_owner_Type":Car_owner_Type,"Price":Price})
driver.quit()
```
